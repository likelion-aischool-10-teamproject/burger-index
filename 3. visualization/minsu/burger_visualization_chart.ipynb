```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from matplotlib import font_manager, rc
import matplotlib
import platform
```

```python
# 한글꺠짐 처리(windows, mac)
path = "C:/Windows/Fonts/malgun.ttf"
if platform.system() == "Windows":
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font', family=font_name)
elif platform.system()=="Darwin":
    rc('font', family='AppleGothic')
else:
    print("Unknown System")
    
# 차트에 마이너스 부분이 깨질때 처리
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
in_path = '../../1. web_crawling+preprocessing/input/'
out_path = '../../1. web_crawling+preprocessing/output/'
```

```python
# 각 브랜드별 DataFrame 생성
burger_df = pd.read_excel(out_path+'burger_df.xlsx')
burger_df.head()
```

```python
burger_df.shape
```

```python
# 긱 브랜드별 지점수 비교
# sns.set(font_scale = 1)
ax = sns.countplot('브랜드', data=burger_df)
ax.set_title('브랜드별 햄버거 지점수')
ax.set_xlabel('브랜드')
ax.set_ylabel('지점수')
plt.show()
# 각 브랜드별 버거킹과 맥도날드의 지점수는 비슷하며, KFC는 가장 적은 지점을 보유하고 있다.
```

```python
# 각 지역별 브랜드 비교
plt.figure(figsize=(16, 8))
plt.xticks(rotation = -45 )
# sns.set(font_scale = 3)
ax = sns.countplot('주소1', data=burger_df)
ax.set_title('지역별 햄버거 지점수')
ax.set_xlabel('지역')
ax.set_ylabel('지점수')
plt.show()
```

```python
# 각 지역별 브랜드 비교
plt.figure(figsize=(16, 8))
plt.xticks(rotation = -45)
ax = sns.countplot('주소1',hue='브랜드', data=burger_df)
ax.set_title('지역별 햄버거 지점수')
ax.set_xlabel('지역')
ax.set_ylabel('지점수')
plt.show()
```
